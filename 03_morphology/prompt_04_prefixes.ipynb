## Prefix distributions (first 2–3 letters)

**Category:** Morphology (proxy)

**Question:** Are prefix strings Zipf-like, and which look most morphemic?

- **Data:** Brown (or Gutenberg) tokens.
- **Items to Count:** Lowercased prefixes: `tok[:2]` or `tok[:3]` (choose one, justify).

**Steps**
- Filter to alphabetic tokens; lowercase.
- Extract 2- or 3-letter prefixes.
- Run Zipf report/plot.
- Interpret: true prefixes (re-, un-, dis-) vs coincidences (th-, co-).

**Deliverable:** Top-15 prefixes + 2 limitations + one proposed “better proxy.”

```python
from nltk.corpus import brown
from nltk import FreqDist
```

```python
# Helper functions
def zipf_prep(items, min_count=1):
    fd = FreqDist(items)
    # drop ultra-rare if desired
    items2 = [(k,v) for k,v in fd.items() if v >= min_count]
    items2.sort(key=lambda kv: kv[1], reverse=True)
    return items2  # list of (item, count) sorted

def zipf_quick_report(items, title="", topn=15):
    ranked = zipf_prep(items)
    print(f"\n=== {title} ===")
    print(f"Prefix Types: {len(ranked)} | Prefixes: {sum(c for _,c in ranked)}")
    print("Top prefixes:")
    for i,(k,v) in enumerate(ranked[:topn], start=1):
        print(f"{i:>2}. {str(k)[:40]:<40} {v}")
    # simple “Zipf-ness” sanity check (no regression): f1 ~ 2*f2 ~ 3*f3?
    if len(ranked) >= 5:
        f1,f2,f3,f4,f5 = [ranked[i][1] for i in range(5)]
        print("\nHeuristic ratios (closer to 1 is more Zipf-like):")
        print(" f1/(2*f2) =", round(f1/(2*f2), 3))
        print(" f1/(3*f3) =", round(f1/(3*f3), 3))
        print(" f1/(4*f4) =", round(f1/(4*f4), 3))
        print(" f1/(5*f5) =", round(f1/(5*f5), 3))
    return ranked

# Optional plotting cell (instructor can pre-enable if matplotlib is available)
def zipf_plot(ranked, n=1000, title="Top 1000 _______"):
    import matplotlib.pyplot as plt
    ranks = list(range(1, min(n, len(ranked))+1))
    freqs = [ranked[i-1][1] for i in ranks]
    plt.figure(figsize=(6,4))
    plt.loglog(ranks, freqs, marker='.')
    plt.xlabel("rank (log)")
    plt.ylabel("frequency (log)")
    plt.title(title)
    plt.tight_layout()
    plt.show()
```

```python
# Load brown tokens
tokens = brown.words()
print(f"Loaded {len(tokens)} tokens from brown")
```

```python
# Lowercased prefixes with words (first 2 or 3 letters)
prefixes = [
    token.lower()[:2] 
    for token in tokens
    if token.isalpha()
]
```

```python
ranked = zipf_quick_report(prefixes,title="Brown Two-Letter Prefix Types")
zipf_plot(ranked,title="Zipf Plot: Brown Two-Letter Prefix Types")
```

### Interpretation

The distribution of prefix strings follows a Zipf-like pattern, with a small number of two-letter prefix strings occurring much more frequently than the rest. Common strings such as `re` and `un` often correspond to real English prefixes, while others such as `th` or `co` may simply be common initial letter sequences. Because this method only extracts the first two letters of each word, it cannot reliably distinguish true prefixes from orthographic patterns.

**Limitations**

1. This method extracts only the first two letters of each word, so many results are orthographic strings rather than genuine prefixes.
2. Some English prefixes are longer than two letters (for example, `inter-`, `under-`, or `trans-`), so they cannot be identified accurately using this approach.

**Better proxy**

A better proxy would be to use a morphological analyzer to identify true prefixes instead of simply extracting the first two letters of each word.
