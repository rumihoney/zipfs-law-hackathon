## LING 498 Hackathon - Suffixes as morpheme proxies (last 3 letters)

**Question: Do suffix strings behave Zipf-like, and do they hint at productive morphology?**

- Data: Brown (or Gutenberg) tokens.
- Items to count: Lowercased suffixes: tok[-3:] for alphabetic words with length ≥ 4.

Steps:
- Keep only alphabetic tokens ([A-Za-z]+).
- Lowercase; filter short words.
- Extract suffixes and run Zipf report/plot.
- Interpret: inflectional-like vs accidental orthographic endings.

Deliverable:
Top-15 suffixes + short claim about productivity vs artifact + 2 limitations.

```python
from nltk.corpus import gutenberg
from nltk import FreqDist

# if NLTK gutenberg corpus isn't already downloaded in runtime, uncomment:
# nltk.download('gutenberg')
```

```python
# Helpter functions
def zipf_prep(items, min_count=1):
    fd = FreqDist(items)
    # drop ultra-rare if desired
    items2 = [(k,v) for k,v in fd.items() if v >= min_count]
    items2.sort(key=lambda kv: kv[1], reverse=True)
    return items2  # list of (item, count) sorted

def zipf_quick_report(items, title="", topn=15):
    ranked = zipf_prep(items)
    print(f"\n=== {title} ===")
    print(f"Types: {len(ranked)} | Tokens: {sum(c for _,c in ranked)}")
    print("Top items:")
    for i,(k,v) in enumerate(ranked[:topn], start=1):
        print(f"{i:>2}. {str(k)[:40]:<40} {v}")
    # simple “Zipf-ness” sanity check (no regression): f1 ~ 2*f2 ~ 3*f3?
    if len(ranked) >= 5:
        f1,f2,f3,f4,f5 = [ranked[i][1] for i in range(5)]
        print("\nHeuristic ratios (closer to 1 is more Zipf-like):")
        print(" f1/(2*f2) =", round(f1/(2*f2), 3))
        print(" f1/(3*f3) =", round(f1/(3*f3), 3))
        print(" f1/(4*f4) =", round(f1/(4*f4), 3))
        print(" f1/(5*f5) =", round(f1/(5*f5), 3))
    return ranked

# Optional plotting cell (instructor can pre-enable if matplotlib is available)
def zipf_plot(ranked, n=1000, title="Top 1000 _______"):
    import matplotlib.pyplot as plt
    ranks = list(range(1, min(n, len(ranked))+1))
    freqs = [ranked[i-1][1] for i in ranks]
    plt.figure(figsize=(6,4))
    plt.loglog(ranks, freqs, marker='.')
    plt.xlabel("rank (log)")
    plt.ylabel("frequency (log)")
    plt.title(title)
    plt.tight_layout()
    plt.show()
```

```python
# Load all gutenberg tokens
tokens = gutenberg.words()
print(f"Loaded {len(tokens)} tokens from gutenberg")
```

```python
# Lowercased suffixes with words (last 3 letters) that have at least 4 letters 
suffixes = [
    token.lower()[-3:] 
    for token in tokens
    if token.isalpha() and len(token) >= 4 
]
```

```python
ranked = zipf_quick_report(suffixes, title="Gutenberg 3-Letter Suffixes")
zipf_plot(ranked, title="Gutenberg 3-Letter Suffixes")
```

## Interpretation

The distribution of three-letter suffixes follows a Zipf-like pattern, with a small number of endings occurring much more frequently than the rest. Some frequent endings, such as `-ing`, reflect productive English morphology because they are associated with common grammatical or derivational patterns. Other common endings are likely accidental orthographic sequences rather than true morphemes, since this method extracts fixed three-letter strings rather than performing morphological analysis.


### Short claim

The results show that extracting the last three letters can identify some real suffixes, but it also captures many endings that are simply common spellings rather than meaningful parts of words.


### Limitations

1. Extracting only the final three letters does not reliably identify true suffixes because English suffixes vary in length and may be attached to different stems.

2. The Gutenberg corpus consists primarily of literary texts, so the frequency of particular endings may reflect the vocabulary and writing style of those works rather than general English usage.
